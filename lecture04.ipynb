# 6 ベクトル空間

## ベクトル空間の公理
## 一次結合
1次結合とは、$c_{i}\in\mathbb{R}$(or $\mathbb{C}$)と$\textbf{a}_{i}\in V$で
$$
\sum_{i=1}^{n}c_{i}\textbf{a}_{i}
$$
で表現したもの。

基底を一次結合に用いると任意のベクトル$\textbf{x}\in V$は
$$
\textbf{x}=\left( 
\begin{array}{c}
x_1 \\
x_2 \\
... \\
x_n \\
\end{array} 
\right)=\sum_{i=1}^{n}x_{i}\textbf{e}_{i}
$$

ただし、(内積を使って直交基底を定義するときにはこのように表現します)
$$
\textbf{e}_{i}\cdot\textbf{e}_{j}=\delta_{ij}
$$

成分を抜き出すときは
$$
\textbf{e}_{j}\cdot\textbf{x}=\textbf{e}_{j}\cdot\sum_{i=1}^{n}x_{i}\textbf{e}_{i}=\sum_{i=1}^{n}x_{i}\textbf{e}_{j}\cdot\textbf{e}_{i}=\sum_{i=1}^{n}x_{i}\delta_{ji}=x_{j}
$$

## 部分空間

### 定理6.2.2(定理というより定義です)

+ $V$を$\mathbb{R}^{r}$上のベクトル空間、$\textbf{a}_i \in\mathbb{R}^{r}$とするとき、

$$
W=\left\{\sum_{i=1}^{r}x_i\textbf{a}_{i}\quad\middle|\quad x_{i}\in\mathbf{R},i=1,...,r\right\}
$$

なる$W$は$V$の線形部分空間(Linear subspace)であり、$W\subset V$を書く。

また、$W$は$\textbf{a}_i \in\mathbb{R}^{r}$によって「張られる(生成される)」ため、$S[\textbf{a}_1,\textbf{a}_2,...\textbf{a}_r]$と表わす。

### 定理6.2.2をいいかえるなら

+ 上記の$W$はそれ単体としてベクトル空間をなす。

### 例

+ $\mathbb{R}_n$や$\mathbb{C}n$に対し、原点を含む 直線、平面、超平面は、全て線型部分空間である。([線型部分空間 \- Wikipedia](https://ja.wikipedia.org/wiki/%E7%B7%9A%E5%9E%8B%E9%83%A8%E5%88%86%E7%A9%BA%E9%96%93))

+ たとえば2次元実ベクトル空間$\mathbb{R}^{2}$、$\textbf{x}_{1},\textbf{x}_{2}\in \mathbb{R}^{2}$で、原点を通る直線$V'_{1}$は
    + $\mathbb{R}^{2}$に比べて$V'_{1}$の次元は1つ落ちているが、表現するには2つのベクトルが必要で、やはり$S[\textbf{x}_{1},\textbf{x}_{2}]$
        + つまり、$\textbf{x}_{1},\textbf{x}_{2}$によって、直線が「張られている」
    + 部分空間は「原点を通る直線」だけでなく、「原点を含む平面」も考えられる。

+ 同じように3次元では、直線と平面、空間が、$\mathbb{R}^{3}$の部分空間として挙げられる。

## 像空間と核空間(p.121)

### 定義

$V$、$W$をベクトル空間とし、線形写像$f:V\longrightarrow W$に対し、

+ 像空間$\mathrm{Im}f$を

$$
\mathrm{Im}f=\left\{f(\textbf{x})\subset W\quad\middle|\quad\textbf{x} \in V \right\}
$$

+ また核空間(Kernel)を

$$
\mathrm{ker}f=\left\{\textbf{x} \in V \quad\middle|\quad f(\textbf{x})=0\right\}
$$

と定義する。

当然ながら、(定理6.3.2)

+ $\mathrm{Im}f\subset W$
+ $\mathrm{ker}f\subset V$

### 例題 6.3.1(p.122)

$$
f:\mathbb{R}^{3}\longrightarrow\mathbb{R}^{3},\quad f\left(\begin{matrix}x\\y\\z\end{matrix}\right)
=\left(\begin{matrix}2x+z\\-x+y\\x+y+z\end{matrix}\right)
$$

定義から像空間は
$$
\mathrm{Im}f=\left\{f\left(\begin{matrix}2x+z\\-x+y\\x+y+z\end{matrix}\right)\in\mathbb{R}^{3}
\middle|\left(\begin{matrix}x\\y\\z\end{matrix}\right)\in\mathbb{R}^{3}\right\}
=\left\{f\left(\begin{matrix}u\\v\\w\end{matrix}\right)\in\mathbb{R}^{3}\middle|w=u+v\right\}
$$

またKernelは
$$
\mathrm{Ker}f=\left\{\left(\begin{matrix}x\\y\\z\end{matrix}\right)\in\mathbb{R}^{3}
\middle|\begin{array}{l}2x+z=0\\-x+y=0\\x+y+z=0\end{array}\right\}
=\left\{\left(\begin{matrix}x\\x\\-2x\end{matrix}\right)\in\mathbb{R}^{3}\quad\middle|\quad x\in\mathbb{R}\right\}
$$


## 6.4 一次独立(一次関係という言葉は特にいらないような気がする)

### 一般形
$\textbf{a}_{i} \in V (i=1,..,n)$が一次独立であるとは、
$$
c_{1}\textbf{a}_{1}+c_{2}\textbf{a}_{2}+...+c_{n}\textbf{a}_{n}=0
$$
が$c_{1}=c_{2}=...=c_{n}=0$のときにしか成り立たない。

#### 一次従属(論理の言葉でいえば「一次独立」の対偶)
$\textbf{a}_{i} \in V (i=1,..,n)$が一次従属であるとは、

$$
c_{1}\textbf{a}_{1}+c_{2}\textbf{a}_{2}+...+c_{n}\textbf{a}_{n}=0
$$

となるのは、$c_{1},c_{2},...,c_{n}$のどれか一つ、もしくは一つ以上が$0$でない。(=少なくとも一つが0でない)

要するに、$\textbf{a}_{i} \in V (i=1,..,n)$を組合せて、ゼロベクトルにすることができる(原点を指すことができる)ということ。



### 2次元で考えます。

$a_{i} \in V (i=1,2)$について
$$
c_{1}\textbf{a}_{1}+c_{2}\textbf{a}_{2}=0
$$
が$c_{1}=c_{2}=0$のときにしか成り立たない。

+ $\textbf{a}_{1}$は$\textbf{a}_{2}$を使っても言い換えができないということ。つまり、

$$
\textbf{a}_{1}=\alpha\textbf{a}_{2}
$$

というような$\alpha \in \mathbb{R}$は存在しえない。

ところで、$\textbf{a}\parallel\textbf{b}$であるとき、スカラー積を用いることで、

$$
\textbf{a}=\alpha\textbf{b}
$$

と表現できる。1次独立のとき$\textbf{a}_{1}\neq\alpha\textbf{a}_{2}$であるから、$\textbf{a}\not\parallel\textbf{b}$となる。
(p.128,"一次独立と1次従属の幾何学的意味(3)")

+ 不要なベクトルを排除して、必要最低限のベクトルで空間を表現したい時に一次独立を定義します。(p.126,コメント)

+ 必ずしも$a_{i} \in V (i=1,2)$のそれぞれが直交していなくていい(p.262, 正規直交基底)

### 例:色の三原色(極端な例、ぱっと思い浮かんだだけなので、間違っているかも)

三原色(青、赤、黄)は、この三色を基底$\textbf{color}_{i}\in \mathbb{R}^{3}$として、$0\le c_i\le1$、$c_i\in\mathbb{R}$を用いて、特徴空間(ベクトル空間)をなし、$\textbf{color}_{i}$のそれぞれは一次独立の関係にある。すなわち、

$$
c_{blue}\textbf{Color}_{blue}+c_{red}\textbf{Color}_{red}+c_{yellow}\textbf{Color}_{yellow}=0,\quad ただし等号はc_i=0のときのみ
$$

+ 要するに「青は赤や黄では表現できない」(ある原色は他の原色を使って表現できない)ということ。
+ $c_i\neq 0$のとき、何らかの色となる(ゼロベクトルならず、どこか原点以外のところを指す)。すなわち、

$$
c_{blue}\textbf{Color}_{blue}+c_{red}\textbf{Color}_{red}+c_{yellow}\textbf{Color}_{yellow}\neq 0,\quad ただしc_i\neq 0
$$


## 6.5 連立斉次1次方程式(p.134)

### せい【斉〔齊〕】の意味

+ 凸凹がなく等しくそろっている。そろえる。ととのえる。
+ 中国、春秋時代の国名。

斉次1次方程式とは、「非零項がすべて同じ次数である」ような方程式。そのような斉次な1次方程式の連立を考える。
$$
\begin{eqnarray}
\left\{
\begin{array}{l}
a_{11}x_{1}+...a_{1n}x_{n} \\
a_{21}x_{1}+...a_{2n}x_{n} \\
..... \\
a_{n1}x_{1}+...a_{nn}x_{n} \\
\end{array}
\right.
\end{eqnarray}
$$

は行列とベクトルを用いることで

$$
A\textbf{x}=0,\quad
A=\left(\begin{array}{ccc}
a_{11} & ... & a_{1n} \\
a_{21} & ... & a_{2n} \\
... & ... & ... \\
a_{n1} & ... & a_{nn} \\
\end{array}
\right),\quad \textbf{x}=\left(\begin{array}{ccc}
x_1 \\
x_2 \\
... \\
x_n \\
\end{array}
\right)
$$

+ 自明な解$\textbf{x}=0$
+ 解空間。すなわち$A\textbf{x}=0$をみたす$\textbf{x}$が張る空間。これはKernelですね。
+ 自明な解以外の解を持つための必要十分条件は$|A|=0$(定理6.5.1)

教科書は丁寧にやっているけれど、僕のテキトー証明は以下。(ちゃんとした証明は教科書を参考ください)

$|A|\neq 0$のとき、$A$は逆行列を持ち、
$$
A\textbf{x}=0 \\
A^{-1}A\textbf{x}=0
$$
となり、
$$
\textbf{x}=0
$$
と自明な解となる。
このようにならないためには、$A$は逆行列をもたない、すなわち$|A|=0$であることである。

イメージは、
$$
\textbf{0}\textbf{x}=0
$$
ように、総体として0になる行列である、という感じ。

演算として、$A\textbf{x}=0$を$\textbf{x}=0$にもっていけないのです。

# 04 行列式-余因子展開(p92)


$A_{ij}\in M(m\times n,\mathbb{R} or \mathbb{C})$なる行列の行列式を考える。

$a_{ij}$における余因子を下記で定義する。

$$
\tilde{a}_{ij}=(-1)^{i+j}\cdot|A_{ij}|
$$

$$
A_{ij}=(-1)^{i+j}\begin{vmatrix}
a_{11}&\cdots&a_{1,j-1}&a_{1,j+1}&\cdots&a_{1n}\\
\vdots&\cdots&\vdots&\vdots&\cdots&\vdots\\
a_{i-1,1}&\cdots&a_{i-1, j-1}&a_{i-1, j+1}&\cdots&a_{i-1, n}\\
a_{i+1,1}&\cdots&a_{i+1, j-1}&a_{i+1, j+1}&\cdots&a_{i+1, n}\\
\vdots&\cdots&\vdots&\vdots&\cdots&\vdots\\
a_{n,1}&\cdots&a_{n, j-1}&a_{n, j+1}&\cdots&a_{n, n}
\end{vmatrix}
$$

+ 第$i$行が除かれている
+ 第$j$行が除かれている

## とりあえずp.93の例をやりましょう

$$
A=\left(
\begin{array}{ccc}
1 & 2 & 0 \\
3 & 5 & -1 \\
4 & 6 & 7 \\
\end{array} 
\right)
$$

の
+ 1行目の余因子$\tilde{a}_{11}$、$\tilde{a}_{12}$、$\tilde{a}_{13}$は?

+ 2行目の余因子$\tilde{a}_{21}$、$\tilde{a}_{22}$、$\tilde{a}_{23}$は?

+ 3行目の余因子$\tilde{a}_{31}$、$\tilde{a}_{32}$、$\tilde{a}_{33}$は?

+ ちなみに行列式$|A|$は?

+ ここまでくると

    + $a_{11}\tilde{a}_{11}+a_{12}\tilde{a}_{12}+a_{13}\tilde{a}_{13}$は?
    + $a_{21}\tilde{a}_{21}+a_{22}\tilde{a}_{22}+a_{23}\tilde{a}_{23}$は?
    + $a_{21}\tilde{a}_{31}+a_{32}\tilde{a}_{32}+a_{33}\tilde{a}_{33}$は?

まとめると、余因子展開を用いて行列式を求めることができる。(定理4.7.1/4.7.2)

$$
\mathrm{det}A=|A|=\sum_{k=1}^{n}a_{ik}\tilde{a}_{ik}=\sum_{l=1}^{n}a_{li}\tilde{a}_{li}
$$
なお、(ここでは証明略。丁寧に計算して確かめるだけです)
$$
a_{i1}\tilde{a}_{k1}+a_{i2}\tilde{a}_{k2}+ ... +a_{in}\tilde{a}_{kn}=\sum_{m=1}^{n}a_{im}\tilde{a}_{km}=0\quad(i\neq k)
$$


## やっと逆行列

### 定理4.7.3

余因子を成分とした行列を転置した行列を余因子行列とよぶ。
$$
\tilde{A}=\left(
\begin{array}{ccc}
\tilde{a}_{11} & .. & \tilde{a}_{1n} \\
.. & .. & .. \\
\tilde{a}_{n1} & .. & \tilde{a}_{nn} \\
\end{array} 
\right)^{T}
=\left(
\begin{array}{ccc}
\tilde{a}_{11} & .. & \tilde{a}_{n1} \\
.. & .. & .. \\
\tilde{a}_{1n} & .. & \tilde{a}_{nn} \\
\end{array} 
\right)
$$

この余因子行列$\tilde{A}$を用いることで、

$$
\mathrm{det}A\cdot E_{n}=A\cdot\tilde{A}=\tilde{A}\cdot A
$$

が成立する。すなわち$A$の逆行列は

$$
A^{-1}=\frac{1}{\mathrm{det}A}\tilde{A}
$$
となる。

#### 証明

$$
[A\cdot\tilde{A}]_{ik}=\sum_{j}a_{ij}\cdot b_{jk}=\sum_{j}a_{ij}\cdot \tilde{a}_{kj}
$$

定理4.7.1 $\mathrm{det}A=|A|=\sum_{k=1}^{n}a_{ik}\tilde{a}_{ik}$より

$$
\sum_{j}a_{ij}\cdot \tilde{a}_{kj}=\mathrm{det}A\cdot \delta_{ik}
$$

すなわち、

$$
[A\cdot\tilde{A}]_{ii}=\mathrm{det}A
$$

となり、

$$
A\cdot\tilde{A}=\mathrm{det}A\cdot E_{n}
$$

また両辺に$A^{-1}$をかけて、さらに$\mathrm{det}A$で割り、右辺から書くと
$$
A^{-1}=\frac{1}{\mathrm{det}A}\cdot\tilde{A}
$$

### 例 問4.7.3 (1)
$$
\left(
\begin{array}{ccc}
1 & 1 & 1 \\
0 & 1 & -1 \\
1 & 1 & 0 \\
\end{array} 
\right)^{-1}
=\frac{1}{\mathrm{det}A}\left(
\begin{array}{ccc}
\tilde{a}_{11} & \tilde{a}_{12} & \tilde{a}_{13} \\
\tilde{a}_{21} & \tilde{a}_{22} & \tilde{a}_{23} \\
\tilde{a}_{31} & \tilde{a}_{32} & \tilde{a}_{33} \\
\end{array} 
\right)^{T}
=\frac{1}{\mathrm{det}A}\left(
\begin{array}{ccc}
\begin{vmatrix} 1 & -1 \\1 & 0 \end{vmatrix} & (-1)\begin{vmatrix} 0 & -1 \\1 & 0 \end{vmatrix} & \begin{vmatrix} 0 & 1\\1 & 1 \end{vmatrix} \\
(-1)\begin{vmatrix} 1 & 1 \\1 & 0 \end{vmatrix} & \begin{vmatrix} 1 & 1 \\1 & 0 \end{vmatrix} & (-1)\begin{vmatrix} 1 & 1\\1 & 1 \end{vmatrix} \\
\begin{vmatrix} 1 & 1 \\1 & -1 \end{vmatrix} & (-1)\begin{vmatrix} 1& 1 \\0 & -1 \end{vmatrix} & \begin{vmatrix} 1 & 1\\0 & 1 \end{vmatrix} 
\end{array} 
\right)^{T} \\
=\frac{1}{(-1)}\left(
\begin{array}{ccc}
1 & -1 & -1 \\
1 & -1 & 0 \\
-2 & 1 & 1 \\
\end{array} 
\right)^{T}
=\left(
\begin{array}{ccc}
1 & 1 & -2 \\
-1 & -1 & 1 \\
-1 & 0 & 1 \\
\end{array} 
\right)
$$

## 正則/逆行列が存在する条件(4.9 正則行列、逆行列 p.101)

+ "逆行列が存在する" = "正則行列である" = "正方行列$A$の行列式$\mathrm{det}A\neq0$"

## 行列式の公式まとめ

$A\in M(n,\mathbb{R} or \mathbb{C})$に対して

1. $\mathrm{det}(cA)=c^{n}\mathrm{det}A,\qquad if A\in\mathbb{R}$

2. $\mathrm{det}(A^{T})=\mathrm{det}(A)$

+ $\mathrm{det}(A\cdot B)=\mathrm{det}(A)\cdot\mathrm{det}(B)$

+ $\mathrm{det}(A^{-1})=\frac{1}{\mathrm{det}(A)}$

+ $\mathrm{det}(A^{n})=(\mathrm{det}A)^{n}$

普通の代数計算とほぼ同じ。証明をしてみて納得するのも大切ですが、実際にパッと書けるようになる方が重要です。
