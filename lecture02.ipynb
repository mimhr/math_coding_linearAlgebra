## たとえば${}^{t}(ABC)={}^{t}C^{t}B^{t}A$の証明

$A,B,C\in M(n,\mathbb{R})$とする。

1.成分での計算。行列の$(i,j)$成分を$(A)_{ij}$のように記述すると

$$\begin{align*}
({}^{t}C{}^{t}B{}^{t}A)_{ij}
=({}^{t}C)_{il}({}^{t}B)_{lm}({}^{t}A)_{mj}
=(C)_{li}(B)_{ml}(A)_{jm}
=(A)_{jm}(B)_{ml}(C)_{li}
=(ABC)_{ji}
={}^{t}(ABC)_{ij}
\end{align*}
$$

2.行列のままでの計算。ここでは${}^{t}(AB)={}^{t}B{}^{t}A$を認めます。
$$
\begin{align*}
{}^{t}(ABC)={}^{t}((AB)C)={}^{t}C {}^{t}(AB)={}^{t} C{}^{t} B{}^{t}A
\end{align*}
$$

## 複素行列

$M(n,\mathbb{C})$と書きます。ただ、この書では、当面$M(n,\mathbb{R})$のみで議論がされます。

### まとめ

$A\in M(n,\mathbb{C})$(ただし、$i,j \in \mathbb{N}$) (必ずしも正方行列である必要はありません)に対して、

複素共役行列とは、「成分の複素共役」をおこなすこと、すなわち、

$$
(\overline{A})_{ij}=\{\overline{a}_{ij}|M(n,\mathbb{C}),i,j\in\mathbb{N}\}
$$

随伴行列とは、「成分の複素共役」と「転置」をおこなすこと、すなわち、
$$
(A^{*})_{ij}=\overline{a}_{ji}
$$

随伴行列が用いられることがほとんどです。というのも、複素数における複素共役は、行列演算における随伴行列にあたります。

### 性質

+ 定理2.7.2(p.39)
$$
\left.
\begin{array}{l}
(A + B)^{*}=A^{*}+B^{*} \\
(\lambda A)^{*}=\overline{\lambda}A^{*} \\
(AB)^{*}=B^{*}A^{*} \\
(A^{*})^{*}=A
\end{array}
\right.
$$

#### $(AB)^{*}=B^{*}A^{*}$の証明
$$
(AB)^{*}_{ij}={}^{t}(\overline{A}_{ik}\overline{B}_{kj})=(\overline{A}_{ki}\overline{B}_{jk})
=\overline{B}_{jk}\cdot\overline{A}_{ki}=B^{*}\cdot A^{*}
$$

### 有名な複素行列

実行列は成分が実数のみ限った複素行列とみることができる。すなわち、$M(n,\mathbb{R})\subset M(n,\mathbb{C})$である。その際、有名な複素行列は以下のように分類できる。

$$
\left.
\begin{array}{lcc}
& M(n,\mathbb{R}) & M(n,\mathbb{C}) \\
転置or随伴したものが同じ\quad A=A^{*} & 対称行列 & Hermite行列 \\
元と転置or随伴との積は単位行列\quad A\cdot A^{*}=A^{*}\cdot A=E_{n}& 直交行列 & Unitary行列 \\
\end{array} 
\right.
$$
